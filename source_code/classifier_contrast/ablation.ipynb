```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb
```

```python
import numpy as np 
import pandas as pd 

df = pd.read_csv('<feature_file>',header=None)
labeldata = pd.read_csv('<label_file>')
Y = labeldata['private']
X = df.iloc[:,:]
```

```python
frequency_features = X.iloc[:, :991]
similarity_features = X.iloc[:, 991:992]
sentiment_features = X.iloc[:, 992:998]
entity_features = X.iloc[:, 998:1014]
semantic_features = X.iloc[:, 1014:]
```

```python
semantic_features.shape
```

```python
from itertools import combinations
```

```python
# choose combinations of features to use for classification
feature_list = [frequency_features,similarity_features, sentiment_features, entity_features, semantic_features]
combinations_list = []
for i in range(2, len(feature_list) + 1):
    for combination in combinations(feature_list, i):
        combinations_list.append(combination)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
import warnings
def warn(*args, **kwargs):
    pass
warnings.warn = warn
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, f1_score, auc,
                              precision_score, recall_score, roc_auc_score, roc_curve, 
                             precision_recall_curve)
```

```python
# using LR
results = []
for combination in combinations_list:
    print('Combination:', [feature.columns[0] for feature in combination])
    Xcom = pd.concat(combination, axis=1)
    print(Xcom.shape)
    
    X_train, X_test, y_train, y_test = train_test_split(Xcom, Y, test_size=0.2, random_state=42)
    
    xgb_classifier = LogisticRegression()

    
    xgb_classifier.fit(X_train, y_train)

   
    y_train_pred = xgb_classifier.predict(X_train)
    y_test_pred = xgb_classifier.predict(X_test)

    
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    class_report = classification_report(y_test, y_test_pred)
    results.append({
        'Combination': [feature.columns[0] for feature in combination],
        'shape':Xcom.shape,
        'Train-Accuracy': train_accuracy,
        'Classification Report': class_report
    })

    # print("Train Accuracy: %.2f%%" % (train_accuracy * 100.0))
    print("Test Accuracy: %.2f%%" % (test_accuracy * 100.0))

# get all combinations results
for result in results:
    print('Combination:', result['Combination'])
    print('Shape:',result['shape'])
    print('Train-Accuracy:', result['Train-Accuracy'])
    print('Classification Report:\n', result['Classification Report'])
    print('\n')
```

```python
# using DNN
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

results = []
Y = Y.values
for combination in combinations_list:
    print('Combination:', [feature.columns[0] for feature in combination])
    Xcom = pd.concat(combination, axis=1)
    print(Xcom.shape)
    st=StandardScaler()
    Xcom= st.fit_transform(Xcom)
   

    X_train, X_test, y_train, y_test = train_test_split(Xcom, Y, test_size=0.2, random_state=42)
    print(X_train.shape)
  
    X_train = torch.FloatTensor(X_train)
    y_train = torch.LongTensor(y_train)
    X_test = torch.FloatTensor(X_test)
    y_test = torch.LongTensor(y_test)

    
    train_dataset = TensorDataset(X_train, y_train)
    train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_dataset = TensorDataset(X_test, y_test)
    test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)

    
    class DNNClassifier(nn.Module):
        def __init__(self, input_dim, hidden_dim, output_dim):
            super(DNNClassifier, self).__init__()
            self.fc1 = nn.Linear(input_dim, hidden_dim)
            self.relu1 = nn.ReLU()
            self.fc2 = nn.Linear(hidden_dim, hidden_dim)
            self.relu2 = nn.ReLU()
            self.fc3 = nn.Linear(hidden_dim, output_dim)

        def forward(self, x):
            out = self.fc1(x)
            out = self.relu1(out)
            out = self.fc2(out)
            out = self.relu2(out)
            out = self.fc3(out)
            return out

   
    model = DNNClassifier(Xcom.shape[1], 20, 2)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    
    criterion = nn.CrossEntropyLoss()

    
    train_losses = []
    test_losses = []
    train_accs = []
    test_accs = []
    for epoch in range(100):
        train_loss = 0
        train_total = 0
        train_correct = 0
        for batch_x, batch_y in train_dataloader:
            optimizer.zero_grad()
            pred_y = model(batch_x)
            loss = criterion(pred_y, batch_y.squeeze())
            loss.backward()
            optimizer.step()

            train_loss += loss.item() * batch_x.shape[0]
            train_total += batch_x.shape[0]
            train_correct += (torch.argmax(pred_y, dim=1) == batch_y.squeeze()).sum().item()

        train_loss /= train_total
        train_acc = train_correct / train_total
        train_losses.append(train_loss)
        train_accs.append(train_acc)

        test_loss = 0
        test_total = 0
        test_correct = 0
        with torch.no_grad():
            for batch_x, batch_y in test_dataloader:
                pred_y = model(batch_x)
                loss = criterion(pred_y, batch_y.squeeze())

                test_loss += loss.item() * batch_x.shape[0]
                test_total += batch_x.shape[0]
                test_correct += (torch.argmax(pred_y, dim=1) == batch_y.squeeze()).sum().item()

        test_loss /= test_total
        test_acc = test_correct / test_total
        test_losses.append(test_loss)
        test_accs.append(test_acc)

        # print('Epoch [{}/{}], Train Loss: {:.4f}, Train Acc: {:.4f}, Test Loss: {:.4f}, Test Acc: {:.4f}'.format(epoch+1, 20, train_loss, train_acc, test_loss, test_acc))
    
    with torch.no_grad():
        y_pred = []
        y_true = []
        for batch_x, batch_y in test_dataloader:
            pred_y = model(batch_x)
            y_pred.extend(torch.argmax(pred_y, dim=1).tolist())
            y_true.extend(batch_y.squeeze().tolist())
        print(classification_report(y_true, y_pred))

      
        fpr, tpr, thresholds = roc_curve(y_true, y_pred)
        roc_auc = auc(fpr, tpr)
        precision = precision_score(y_true, y_pred)
        recall = recall_score(y_true, y_pred)
        f1 = f1_score(y_true, y_pred)
        print('Combination:', result['Combination'])
        print('ROC AUC: {:.4f}'.format(roc_auc))
        print('precision: {:.4f}'.format(roc_auc))
        print('recall: {:.4f}'.format(roc_auc))
        print('f1: {:.4f}'.format(roc_auc))
        print("train loss:",train_losses)
        print("test loss: ",test_losses)
        print("train acc:",train_accs)
        print("test acc:",test_accs)
```
